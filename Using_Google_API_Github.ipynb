```python
#Creating Amenity Maps for Catalogues

All_types=['bank','bar','restaurant','cafe']

List_Searches=[]
for i in All_types:
    places=gmaps.places_nearby(location= '50.963804, -114.073427',radius= 2000,open_now=False,type=i)
    time.sleep(2)
    List_Searches.append(places)

j=0
list_of_places=[]
list_of_latitude=[]
list_of_longitude=[]

for i in range(len(List_Searches)):
    amenity=All_types[j]
    for place in List_Searches[i]['results']:
        my_place_id=place['place_id']
        fields = ['name', 'formatted_address','geometry','international_phone_number', 'website', 'rating','type']
        place_details=gmaps.place(place_id=my_place_id,fields=fields)
        list_of_places.append(place_details['result'])
        list_of_latitude.append(place_details['result']['geometry']['location']['lat'])
        list_of_longitude.append(place_details['result']['geometry']['location']['lng'])
    j+=1    

dfObj=pd.DataFrame(list_of_places)
coordinates_places = pd.DataFrame(zip(list_of_latitude,list_of_longitude))
dffinal=pd.concat([dfObj, coordinates_places], axis=1)

#Time String
timestr = time.strftime("%Y%m%d-%H%M%S")
dffinal.to_csv('Layers_{}.csv'.format(timestr))
        



```
